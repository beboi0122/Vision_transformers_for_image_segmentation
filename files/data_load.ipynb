```python
from PIL import Image
import matplotlib.pyplot as plt
import os

# Set the paths to the dataset
image_path = 'electric_wires_dataset/images'  # Adjust based on extracted folder structure
mask_path = 'electric_wires_dataset/masks'    # Adjust based on extracted folder structure

# Load images and masks
images = []
masks = []

for filename in os.listdir(image_path):
    if filename.endswith('.png'):  # Adjust based on file types
        img = Image.open(os.path.join(image_path, filename))
        images.append(img)

for filename in os.listdir(mask_path):
    if filename.endswith('.png'):  # Adjust based on file types
        mask = Image.open(os.path.join(mask_path, filename))
        masks.append(mask)

# Display an example image and its mask
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Sample Image')
plt.imshow(images[0])  # Display the first image
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Sample Mask')
plt.imshow(masks[0])  # Display the first mask
plt.axis('off')

plt.show()
```
